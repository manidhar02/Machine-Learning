### Task-1 (Reading the given files in Python notebook and storing the variables X, trueLabels, and terms)


```python
#Reading the data from the given csv files
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn import metrics
from scipy import *
import matplotlib.pyplot as plt
import sklearn as SKL
import sys
%matplotlib inline

X = pd.read_csv('bbcsport_mtx.csv', delimiter=',', header=None)
terms = pd.read_csv('bbcsport_terms.csv', delimiter=',', header=None).values
trueLabels = pd.read_csv('bbcsport_classes.csv', delimiter=',', header=None).values
```

```python
#Converting the data to array and displaying it
Feature_Matrix = X.values #.values convert the dataframe to a numpy array
print Feature_Matrix      
print "Shape of matrix is: {}".format(Feature_Matrix.shape)
trueLabels = np.array(trueLabels)  # converting trueLabels to an array
trueLabels = np.array(trueLabels.T)[0]  # Transpoing the matrix 
print "Shape of trueLabels is: {}".format(trueLabels.shape)
```


### Task-2 (Performing K-means clustering with 5 clusters using Euclidean distance as similarity measure and evaluating the clustering performance using adjusted rand index and adjusted mutual information. And finding  the clustering performance averaged over 50 random initializations of K-means):

```python
# Performing K-means clustering using Euclidean distance as similarity measure

Rand_Index = range(50) #creating a range of 50 values 
Adjusted_Mutual_Info = range(50)
for num in range(50):
    kmeans = KMeans(n_clusters=5, init='random',n_init=1) #n_clusters=5 represents the number of clusters to find
    kmeans.fit(Feature_Matrix)     
    centroids = kmeans.cluster_centers_
    labels    = kmeans.labels_
    adjusted_rand_score = metrics.adjusted_rand_score(trueLabels, labels)
    Rand_Index[num] = adjusted_rand_score
    adjusted_mutual_info = SKL.metrics.adjusted_mutual_info_score(trueLabels, labels)
    Adjusted_Mutual_Info[num] = adjusted_mutual_info
print labels
```

```python
# Averaged clustering performance over 50 random initializations of K-means

Avergae_Rand_Index = np.array(Rand_Index)
Avergae_Rand_Index_Value = np.mean(Avergae_Rand_Index, axis=0) # Finding the Mean for ountput of 50 random initializations 
print "The Average adjusted rand index of k-means cluster over 50 random initializations: {}".format(Avergae_Rand_Index_Value)
Average_Adjusted_Mutual_info = np.array(Adjusted_Mutual_Info)
Average_Adjusted_Mutual_info_Value = np.mean(Average_Adjusted_Mutual_info, axis=0)
print "The Average adjusted mutual information of k-means cluster over 50 random initializations: {}".format(Average_Adjusted_Mutual_info_Value)
```

Analysis:
Rand index or Rand measure, computers the similarity between 2 data clustering. We use a modified measure called **adjusted Rand index**.  Though the Rand Index may only yield a value between 0 and +1, the adjusted Rand index can yield negative values if the index is less than the expected index.
   
   Here we got the values in between o and 1 only as expected.



### Task-3 (Performing K-means clustering with 5 clusters using Cosine distance as similarity measure and evaluating the clustering performance using adjusted rand index and adjusted mutual information. And finding  the clustering performance averaged over 50 random initializations of K-means):

```python
# Performing K-means clustering using Cosine distance as similarity measure

Normalised = preprocessing.normalize(Feature_Matrix,norm='l2')  #The norm to use to normalize each non zero sample (or each non-zero feature if axis is 0).
Rand_Index_Cosine = range(50) #creating a range of 50 values 
Adjusted_Mutual_Info_Cosine = range(50)
for num in range(50):
    kmeans = KMeans(n_clusters=5, init='random',n_init=1) #n_clusters=5 represents the number of clusters to find
    kmeans.fit(Normalised)     
    centroids = kmeans.cluster_centers_
    labels_cosine    = kmeans.labels_
    adjusted_rand_score = metrics.adjusted_rand_score(trueLabels, labels_cosine)
    Rand_Index_Cosine[num] = adjusted_rand_score
    adjusted_mutual_info = SKL.metrics.adjusted_mutual_info_score(trueLabels, labels_cosine)
    Adjusted_Mutual_Info_Cosine[num] = adjusted_mutual_info
print labels
```

```python
# Averaged clustering performance over 50 random initializations of K-means

Avergae_Rand_Index = np.array(Rand_Index_Cosine)
Avergae_Rand_Index_Value = np.mean(Avergae_Rand_Index, axis=0) # Finding the Mean for output of 50 random initializations 
print "The Average adjusted rand index of k-means cluster over 50 random initializations: {}".format(Avergae_Rand_Index_Value)
Average_Adjusted_Mutual_info = np.array(Adjusted_Mutual_Info_Cosine)
Average_Adjusted_Mutual_info_Value = np.mean(Average_Adjusted_Mutual_info, axis=0)
print "The Average adjusted mutual information of k-means cluster over 50 random initializations: {}".format(Average_Adjusted_Mutual_info_Value)
```

```python
#Function for adding the feature matrix for each cluster based on the index values
def MatrixAddition(cluster_list):
    j= len(cluster_list)
    cluster = np.zeros([1,4613], dtype=np.int)
    for i in range(len(cluster_list)):
        cluster= cluster+Feature_Matrix[cluster_list[i]] 
    cluster_transpose= cluster.T
    return cluster_transpose
```

```python
#Function for generating the words usig cluster after K means
def wordsGeneration(cluster_transpose):
    x=0
    y=0
    cluster_words = ""
    terms_cluster = ""
    while x < len(cluster_transpose):
        y = cluster_transpose[x]
        if y > 0:
            for p in range(y):
                terms_cluster = terms[x]
                cluster_words = terms_cluster + " " + cluster_words
        x+=1
    return cluster_words
```

```python
#Function for getting the string from words and to create word plot
def wordCloudPlot(cluster_words):
    from wordcloud import WordCloud, STOPWORDS
    cluster_text= np.array2string(cluster_words) 
    bagOfWords = WordCloud(stopwords=STOPWORDS,background_color='black',collocations=False).generate_from_text(cluster_text)
    return bagOfWords
```

### Part-2 (Dimensionality Reduction using PCA/SVD):

```python
#normalize our data
from sklearn.preprocessing import scale
Xnorm = scale(X)
```

```python
#performing PCA using sklearn PCA implementation
from sklearn.decomposition import PCA
pca = PCA(n_components=4613)
pca.fit(Xnorm)
```

```python
#The amount of variance that each PC explains
var= pca.explained_variance_ratio_
print var.shape
```

```python
#Cumulative Variance explains
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
print var1
plt.plot(var1)
plt.xlabel("Principal components")
plt.ylabel("Variance captured")
```

Analysis:
    The minimum dimension that captures (a) at least 95% variance is 555  and (b) at least 98% variance is 626
